### Analysis

- When compared to the placebo, both Capomulin and Ramicane show consistantly lower scores in tumor volume mean, median, variance, standard deviation, and standard error of the mean.
- These same two drug trials also had the highest number of timepoints, suggesting the mice lived longer.
- However, despite these promising numbers, a closer analysis of Capomulin shows that the tumors continue to grow with time. 
 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
mouse_data_complete = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview

mouse_data_complete.head()
```

```python
# Checking the number of mice.
mice_number = len(mouse_data_complete['Mouse ID'].value_counts())
mice_number
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
duplicate_mouse = mouse_data_complete.loc[mouse_data_complete.duplicated(subset = ['Mouse ID', 'Timepoint'], keep=False), 'Mouse ID'].unique()
duplicate_mouse
```

```python
# Optional: Get all the data for the duplicate mouse ID.
removed_mouse = mouse_data_complete[mouse_data_complete['Mouse ID']==duplicate_mouse[0]]
removed_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_mouse=mouse_data_complete[mouse_data_complete['Mouse ID']!=duplicate_mouse[0]]
new_mouse
```

```python
# Checking the number of mice in the clean DataFrame.
new_mouse_number = len(new_mouse['Mouse ID'].value_counts())
new_mouse_number
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen:
# mean, median, variance, standard deviation, and SEM of the tumor volume.
mean_vol=new_mouse.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].mean()
med_vol=new_mouse.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].median()
var_vol=new_mouse.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].var()
stdev_vol=new_mouse.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].std()
sem_vol=new_mouse.groupby(['Drug Regimen'])['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary DataFrame.
drug_summary = pd.DataFrame({'Mean Tumor Volume' : mean_vol, 'Median Tumor Volume' : med_vol, 'Tumor Volume Variance' : var_vol,	
                 'Tumor Volume Std. Dev' : stdev_vol, 'Tumor Volume Std. Err.' : sem_vol})
drug_summary
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
agg_result = new_mouse.groupby(['Drug Regimen']).agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']})
agg_result
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:
# Capomulin, Ramicane, Infubinol, and Ceftamin
#treaments = new_mouse.loc[(new_mouse[
# Start by getting the last (greatest) timepoint for each mouse
last_timepoint = new_mouse.groupby(['Mouse ID'])['Timepoint'].max()
last_timepoint = last_timepoint.reset_index()

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged_timepoint = last_timepoint.merge(new_mouse, on=['Mouse ID', 'Timepoint'], how='left')
merged_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treaments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.
for treatment in treaments:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_tumor_volumes = merged_timepoint.loc[merged_timepoint['Drug Regimen'] == treatment, 'Tumor Volume (mm3)']

    # add subset (I had to look up what this even means. I thought it was a function I had never heard of. This could be worded better I think)
    tumor_vol_data.append(drug_tumor_volumes)

    # Determine outliers using upper and lower bounds
    quartiles = drug_tumor_volumes.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    low_outlier = drug_tumor_volumes < lower_bound
    high_outlier = drug_tumor_volumes > upper_bound
    outliers = drug_tumor_volumes.loc[(low_outlier) | (high_outlier)]
    print(f"{treatment}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distribution of the tumor volume for each treatment group.
fig1, ax1 = plt.subplots()
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(tumor_vol_data, labels = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'], 
            flierprops=dict(markerfacecolor='red', marker='o', markersize = 14))
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin my poor mouse :( 
capomulin_treatment = new_mouse.loc[new_mouse['Drug Regimen'] == 'Capomulin']
single_mouse = capomulin_treatment.loc[new_mouse['Mouse ID'] == 'r157']
plt.plot(single_mouse ['Timepoint'], single_mouse ['Tumor Volume (mm3)'], color="blue")
plt.xlabel('Tumor Volume (mm3)')
plt.ylabel('Timepoint (days)')
plt.title('Capomulin treatment of mouse r157')
plt.show()
# my poor mouse :( 
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
mean_tumor=capomulin_treatment.groupby(['Mouse ID']).mean(numeric_only=True)
plt.scatter(mean_tumor['Weight (g)'],mean_tumor['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
x_values = mean_tumor['Weight (g)']
y_values = mean_tumor['Tumor Volume (mm3)']
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(x_values,regress_values,"r-")
plt.scatter(mean_tumor['Weight (g)'],mean_tumor['Tumor Volume (mm3)'])
plt.xlabel('Weight (g)')
plt.ylabel('Average Tumor Volume (mm3)')

weight = mean_tumor.iloc[:,1]
volume = mean_tumor.iloc[:,3]
correlation = st.pearsonr(weight,volume)
print(f'The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}')
```
